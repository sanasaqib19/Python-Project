```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 


#Read the data

df = pd.read_csv(r'C:\Users\sanas\Downloads\movie dataset\movies.csv')

```

```python
df.head()
```

```python
#check for missing data

for col in df.columns:
    percentMissing = np.mean(df[col].isnull())
    percentMissingRounded = round(percentMissing,1)
    print('{}-{}%'.format(col,percentMissingRounded))
```

```python
#drop missing data

df.dropna(inplace=True)

for col in df.columns:
    percentMissing = np.mean(df[col].isnull())
    percentMissingRounded = round(percentMissing,1)
    print('{}-{}%'.format(col,percentMissingRounded))
```

```python
# looking at dfata types of each column

df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head(5)
```

```python
#year is not always equal to release date year

#let's first delete location from released column

df['released']=df['released'].str[-4:]

```

```python
dfCopy = pd.read_csv(r'C:\Users\sanas\Downloads\movie dataset\movies.csv')

df['released'] = dfCopy['released']

df.head(3)
```

```python
df['releasedEdit']=df['released'].str[-6:]

df.head(3)
```

```python
df = df.drop('released',1)
df.head(3)
```

```python
df['yearEdit'] = df['releasedEdit']

df.head(3)
```

```python
df = df.drop('releasedEdit',1)
df.head(3)
```

```python
df['votes'] = df['votes'].astype('int64')
df.head(3)
```

```python
#order the data by highest gross revenue

df = df.sort_values(by=['gross'],inplace = False, ascending = False)

df.head()
```

```python
#budget and company size correlation to gross 

```

```python
# scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title ('Budget Vs Gross Earning Correlation')

plt.xlabel('Movie Budget/billion')

plt.ylabel('Gross Earnings/100 Million')

plt.show()
```

```python
#Regplot Budget vs Gross

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
corrMatrix = df.corr(method='pearson')

sns.heatmap(corrMatrix, annot=True)

plt.title('Correlation Matrix')

plt.show()

```

```python
#budget to gross has a correlation of 0.74. So it is Highly correlated
```

```python
#numerize company name so we can use it for correlation 

dfNum = df

for col in dfNum.columns:
    if (dfNum[col].dtype == 'object'):
        dfNum[col] = dfNum[col].astype('category')
        dfNum[col] = dfNum[col].cat.codes
        
dfNum.head()
```

```python
corrMatrix = dfNum.corr(method='pearson')

sns.heatmap(corrMatrix, annot=True)

plt.title('Correlation Matrix')

plt.show()
```
